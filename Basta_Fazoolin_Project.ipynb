```python
class Business:
    
    def __init__(self, name, franchises):
        self.name = name
        self.franchises = franchises
  
    def __repr__(self):
        return str(self.franchises)

franchise_list = ["flagship_store", "new_installment"]

basta = Business("Basta Fazoolin' with my Heart", franchise_list)

#print(basta.franchises)
#print(basta.name)

# arepa details:
# arepe Menu Items
arepa_items = {
  'arepa pabellon': 7.00, 'pernil arepa': 8.50, 'guayanes arepa': 8.00, 'jamon arepa': 7.50
}

#############################################################################################################
#creating a class (object) and have all the attributes that make up that object 
# using the __init__() constructer. Below we have constructed a Franchise object that will
# represent any Franchise we create. 
# The attributes in Franchise ,will be constructed with are as follows:
# address - address of where the Franchise is located
# menus - all the available Menus - menus will consist of a list of all the menus as follows: 
#  + branch_menu
#  + early_bird_menu
#  + dinner_menu
#  + kids_menu

class Franchise:
    
    def __init__(self, address, menus):
        self.address = address
        self.menus = menus

    def __repr__(self):
        return ("Franchise store at " + self.address)

    def available_menus(self, time):
        available_menu_list = []
        for menu in self.menus:
            
            #print(time)
            #print(self.menus)
            #print(menu.start_time)
            #print(menu.end_time)
            if time >= menu.start_time and time <= menu.end_time:
                
                available_menu_list.append(menu)
                
        return available_menu_list
    
    
##########################################################################################################
# creating a class (object) and have all the attributes that make up that object 
# using the __init__() constructer. Below we have constructed a Menu object that will
# represent any Menu we create. 
# The attributes that this Menu ,will be constructed with are as follows:
# name - Name of the Menu
# items - The menu items (aka food choices) which will be a dictionary of the food items and their price
# start_time - represented in 24 hour clock
# end_time - represented in 24 hour clock


class Menu:
    
    def __init__(self, name, items, start_time, end_time):
        
        self.name = name
        self.items = items
        self.start_time = start_time
        self.end_time = end_time
    
    
# REPRESENTATION - uses __repr__(). Returns a STRING value of description of the Menu 

    def __repr__(self):
        return (self.name + " Menu, available from " + str(self.start_time) + " till " + str(self.end_time))
    
    def calculate_bill(self, purchased_items):
        bill_total = 0
        for item in purchased_items:
          bill_total += self.items[item]

        return (self.name + " total: " + str(bill_total)) 

# Below are all the menus - in dictionary format - keys = food : values = price - e.g. bacon & eggs : 10.50

# Brunch Menu Items 
brunch_items = {
  'pancakes': 7.50, 'waffles': 9.00, 'burger': 11.00, 'home fries': 4.50, 'coffee': 1.50, 'espresso': 3.00, 'tea': 1.00, 'mimosa': 10.50, 'orange juice': 3.50
}

# Early Bird Menu Items
early_bird_items = {
  'salumeria plate': 8.00, 'salad and breadsticks (serves 2, no refills)': 14.00, 'pizza with quattro formaggi': 9.00, 'duck ragu': 17.50, 'mushroom ravioli (vegan)': 13.50, 'coffee': 1.50, 'espresso': 3.00,
}

# Dinner Menu Items
dinner_items = {
  'crostini with eggplant caponata': 13.00, 'caesar salad': 16.00, 'pizza with quattro formaggi': 11.00, 'duck ragu': 19.50, 'mushroom ravioli (vegan)': 13.50, 'coffee': 2.00, 'espresso': 3.00,
}

# Kids Menu Items
kids_items = {
  'chicken nuggets': 6.50, 'fusilli with wild mushrooms': 12.00, 'apple juice': 3.00
}


brunch_menu = Menu("Brunch", brunch_items, 1100, 1600)
early_bird_menu = Menu("Early Bird", early_bird_items, 1500, 1800)
dinner_menu = Menu("Dinner", dinner_items, 1700, 2300)
kids_menu = Menu("Kids", kids_items, 1100, 2100)



#
#print(early_bird_menu)
#for item, price in dinner_menu.items.items():
#    print(item, price)


# create the list of menus 
menus = [brunch_menu, early_bird_menu, dinner_menu, kids_menu]

flagship_store = Franchise("1232 West End Road", menus)
new_installment = Franchise("12 East Mulberry Street", menus)

#print(new_installment.menus)

#print(flagship_store)

print(flagship_store.available_menus(1700))


# Arepas Menu
arepas_menu = Menu("Take aâ€™ Arepa", arepa_items, 1000, 2000)

# create the first "Take aâ€™ Arepa" franchise
arepas_place = Franchise("189 Fitzgerald Avenue", arepas_menu)

#make new Business called "Take a' Arepa"
arepa = Business("Take a' Arepa", arepas_place)



print(arepas_menu.items)
print(basta.franchises)
```
